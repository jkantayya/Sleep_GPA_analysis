```python
def sampling_dist(reps = 5000, n1 = df_study1.shape[0], n5 = df_study5.shape[0], withReplace = True, var = "cum_gpa"):
    sim_dist = []
    for i in range(reps):
        sample1 = df_study1[var].sample(n1, replace = withReplace)
        sample2 = df_study5[var].sample(n5, replace = withReplace)
        med_diff = sample1.median() - sample2.median()
        sim_dist.append(med_diff)
    sim_dist = pd.DataFrame({"Median Difference (2018 CMU - 2017 CMU)": sim_dist})
    return sim_dist
```
