```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("cmu-sleep.csv")
df
```

```python
df.demo_firstgen.unique()
```

```python
df.shape
```

```python
# cleaning the firstgen column
df['demo_firstgen'] = df['demo_firstgen'].replace([' ', '2'], np.nan)

df = df.dropna(subset=['demo_firstgen'])
df.shape
```

```python
df.demo_firstgen.unique()
```

```python
sns.boxplot(data=df, x='demo_firstgen', y='term_gpa', palette="Set2")
plt.xticks([0, 1], ['Non-First Gen', 'First Gen'])
plt.title("Term GPA by First-Generation Status")
plt.xlabel("First-Generation Status")
plt.ylabel("Term GPA")
plt.ylim(2, 4.0)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df[df['demo_firstgen'] == '0']['term_gpa'], kde=True, color='blue', label='Non-First Gen', stat="density", linewidth=0)
sns.histplot(df[df['demo_firstgen'] == '1']['term_gpa'], kde=True, color='orange', label='First Gen', stat="density", linewidth=0)
plt.legend()
plt.title("Term GPA Distribution by First-Generation Status")
plt.xlabel("Term GPA")
plt.ylabel("Density")
plt.show()
```

```python
df['demo_gender'] = df['demo_gender'].replace([' ','2'], np.nan)

df = df.dropna(subset=['demo_gender'])
```

```python
sns.boxplot(data=df, x='demo_gender', y='term_gpa', palette="Set2")
plt.xticks([0, 1], ['Male', 'Female'])
plt.title("Term GPA by Gender")
plt.xlabel("Gender")
plt.ylabel("Term GPA")
plt.ylim(2, 4.0)
plt.show()
```
